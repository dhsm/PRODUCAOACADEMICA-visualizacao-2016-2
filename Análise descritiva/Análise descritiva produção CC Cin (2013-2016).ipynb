```python
def get_research_method(title):
    if "CASE" in title:
        return "applied"
    if "SURVEY"in title or "EVALUATION" in title or "STUDY"in title:
        return "survey"
    if "APPROACH" in title or "FRAMEWORK" in title or "ALGORITHM" in title:
        return "pure"
```

```python
df['research_method'] = df['name'].apply(get_research_method)
```
